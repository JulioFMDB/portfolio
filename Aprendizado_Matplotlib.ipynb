```python
fig = plt.figure(figsize=(8, 3));
eixo = fig.add_axes([0,0,1,1]); #criando o gráfico do zero
eixo.set_title('Temperatura no tempo', fontsize=24, loc='right');
eixo.set_xlabel('Data');
eixo.set_ylabel('Temperatura');
eixo.set_xlim(datetime.datetime(2016,1,1), datetime.datetime(2016,12,31))
eixo.grid(True)
eixo.plot(dados['data'], dados['temperatura'], color='#44EB76', lw=.3, ls = 'dotted', marker = '.'); 

eixo2 = fig.add_axes([0.697, .69, .3, .3])
eixo2.set(yticklabels=[])
eixo2.set_xlabel('Temperatura no tempo', fontsize=10);
# eixo2.grid(True)

serie_left = pd.to_datetime(dados['data']) < datetime.datetime(2016,1,1)
serie_right = pd.to_datetime(dados['data']) > datetime.datetime(2017,1,1)



eixo2.plot(dados['data'], dados['temperatura'], color='#44EB76', lw=.1); 
eixo2.plot(dados[serie_left]['data'], dados[serie_left]['temperatura'], color='#1E9E44', lw=.2);
eixo2.plot(dados[serie_right]['data'], dados[serie_right]['temperatura'], color='#1E9E44', lw=.2);

eixo2.legend(['temperatura'], loc='best', fontsize=8)
# eixo2.set_xticks()
```
